### Crea un mapa interactivo con la información de tu diccionario

```python
# Cargamos folium en la memoria de Python para poder usar sus mapas y herramientas
import folium

# Inicializamos el objeto mapa. Centrado en el planeta y con zoom panorámico (escala 2)
mapa_cine = folium.Map(location=[20.0, 0.0], zoom_start=2)
```

```python
# Arrancamos el bucle 'for' para desglosar el diccionario principal fila por fila
# 'clave' tomará valores como "item_1", "item_2"...
# 'info' tomará los sub-diccionarios internos con los datos de las películas
for clave, info in top_peliculas_series.items():

    # Diseñamos una plantilla de texto estructurado en lenguaje HTML.
    # Las etiquetas <strong> hacen que los títulos se vean resaltados en negrita.
    # Las etiquetas <br> obligan a hacer un salto de línea para que no se vea todo amontonado.
    # Llamamos a los datos internos usando la sintaxis: info['nombre_de_la_clave']
    texto_popup = f"""
    <strong>Película/Serie:</strong> {info['nombre']}<br>
    <strong>Director:</strong> {info['director']}<br>
    <strong>Año:</strong> {info['anio']}<br>
    <strong>Género:</strong> {info['genero']}<br>
    <strong>Latitud:</strong> {info['latitud']}<br>
    <strong>Longitud:</strong> {info['longitud']}
    """

    # Configuramos el pin (marcador geográfico) para la película actual del bucle
    marcador_actual = folium.Marker(
        location=[info['latitud'], info['longitud']],          # Ubica el pin usando las coordenadas del diccionario
        popup=folium.Popup(texto_popup, max_width=300),         # Añade la cajita de datos HTML y define su ancho máximo
        tooltip=info['nombre'],                                  # Muestra un minitexto flotante cuando pasas el cursor encima
        icon=folium.Icon(color="red", icon="film", prefix="fa")  # Dibuja un icono rojo con forma de cinta de película
    )

    # Añadimos físicamente el marcador que acabamos de configurar dentro de nuestro mapa base
    marcador_actual.add_to(mapa_cine)

# Al escribir el nombre del mapa al final, obligamos a Jupyter/Colab a renderizar el mapa interactivo
mapa_cine
```
