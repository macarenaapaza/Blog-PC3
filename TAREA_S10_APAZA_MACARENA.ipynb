```python
# Paso 1 y 2: Abrir el archivo de texto y asignamos su contenido

with open('harry_potter_1.txt', 'r', encoding='utf-8') as archivo:
    contenido = archivo.read()

# Verificamos los primeros caracteres para confirmar que se leyó correctamente

print("Archivo cargado con éxito. Primeros 100 caracteres:")
print(contenido[:100])
```

```python
# Paso 3: Generar una lista de palabras a partir del contenido original

lista_palabras_inicial = contenido.split()
print(f"Cantidad de palabras iniciales: {len(lista_palabras_inicial)}")
```

```python
# Paso 4: Convertir a minúsculas y eliminar signos de puntuación y saltos de línea
lista_limpia = []
signos_a_quitar = ",.¿?¡!();:«»\"'-—" # Lista de signos comunes

for palabra in lista_palabras_inicial:
    # Pasamos a minúsculas y quitamos saltos de línea
    palabra_procesada = palabra.lower().replace('\n', '')

    # Quitamos los signos de puntuación uno por uno
    for signo in signos_a_quitar:
        palabra_procesada = palabra_procesada.replace(signo, '')

    # Si la palabra no quedó vacía tras quitarle los signos, la agregamos
    if palabra_procesada != '':
        lista_limpia.append(palabra_procesada)

print(f"Cantidad de palabras después de limpiar signos: {len(lista_limpia)}")
```

```python
# Paso 5: Definir las stopwords en español y eliminarlas de la lista
stopwords_es = [
    'el', 'la', 'los', 'las', 'un', 'una', 'unos', 'unas', 'y', 'e', 'o', 'u',
    'de', 'del', 'a', 'al', 'en', 'con', 'por', 'para', 'que', 'no', 'si', 'se',
    'su', 'sus', 'lo', 'le', 'les', 'me', 'mi', 'te', 'como', 'pero', 'mas'
]

lista_depurada = []
for palabra in lista_limpia:
    if palabra not in stopwords_es:
        lista_depurada.append(palabra)

print(f"Cantidad de palabras en la lista depurada: {len(lista_depurada)}")
```

```python
# Paso 6: Crea un diccionario vacío para almacenar las frecuencias
frecuencia_palabras = {}

for palabra in lista_depurada:
    if palabra in frecuencia_palabras:
        frecuencia_palabras[palabra] += 1
    else:
        frecuencia_palabras[palabra] = 1

# Paso 7: Imprimir el diccionario y verificar la suma
print("Diccionario de frecuencias:")
print(frecuencia_palabras)
print("-" * 50)

total_lista = len(lista_depurada)
suma_valores = sum(frecuencia_palabras.values())

print(f"Total de palabras en la lista: {total_lista}")
print(f"Suma de valores en el diccionario: {suma_valores}")
print(f"¿Coinciden los totales?: {total_lista == suma_valores}")
```

```python
# Paso 8: Crear una nube de palabras con el diccionario obtenido
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Personalizamos con fondo blanco y paleta 'verdes'
nube = WordCloud(
    width=800,
    height=400,
    background_color='white',
    colormap='viridis'
).generate_from_frequencies(frecuencia_palabras)

plt.figure(figsize=(10, 5))
plt.imshow(nube, interpolation='bilinear')
plt.axis('off')  # Quitamos los ejes del gráfico
plt.show()
```

```python
# Paso 9: Indicar cuántas veces se repiten "Dursley", "Potter" y "Dudley"
# Como todo está en minúsculas en el diccionario, buscamos en minúsculas
palabras_buscar = ['dursley', 'potter', 'dudley']

print("Resultado de la búsqueda:")
for p in palabras_buscar:
    conteo = frecuencia_palabras.get(p, 0)
    print(f"La palabra '{p.capitalize()}' se repite: {conteo} veces.")
```
